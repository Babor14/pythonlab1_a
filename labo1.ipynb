```python
#variant5
import numpy as np

def has_equal_positive_negative_column(matrix):
    arr = np.array(matrix)
    num_columns = arr.shape[1]
    
    for col in range(num_columns):
        positive_count = np.sum(arr[:, col] > 0)
        negative_count = np.sum(arr[:, col] < 0)
        
        if positive_count == negative_count:
            return True 
    
    return False  


matrix = [
    [1, -1, 2],
    [2, 1, -2],
    [-1, -2, 3],
    [0, 0, 0]
]

result = has_equal_positive_negative_column(matrix)
print("есть?:", result)
```

```python
def count_greater_than_neighbors(numbers):
    count = 0
   
    for i in range(1, len(numbers) - 1):
        if numbers[i] > numbers[i - 1] and numbers[i] > numbers[i + 1]:
            count += 1
    return count


numbers = [1, 3, 2, 5, 4, 6, 2, 8]
result = count_greater_than_neighbors(numbers)
print("Количество элементов, которые больше двух своих соседей:", result)
```

```python
def count_common_colors(set_a, set_b):
    # Преобразуем списки в множества для удобства
    colors_a = set(set_a)
    colors_b = set(set_b)
    
    # Находим пересечение множеств
    common_colors = colors_a.intersection(colors_b)
    
    # Возвращаем количество общих цветов
    return len(common_colors)

# Пример использования
set_a = [0, 1, 2, 3, 4, 5]  # Набор Ани
set_b = [3, 4, 5, 6, 7, 8]  # Набор Бори

result = count_common_colors(set_a, set_b)
print("Количество общих цветов кубиков:", result)
```

```python
def main():
  
    N, M = map(int, input().split())
    
    
    colors_a = {int(input().strip()) for _ in range(N)}
    colors_b = {int(input().strip()) for _ in range(M)}
    
   
    common_colors = colors_a.intersection(colors_b)
    only_a = colors_a.difference(colors_b)
    only_b = colors_b.difference(colors_a)
    
    # из множества в списки
    common_colors = sorted(common_colors)
    only_a = sorted(only_a)
    only_b = sorted(only_b)
    
   
    print(len(common_colors), *common_colors)
    print(len(only_a), *only_a)
    print(len(only_b), *only_b)


if __name__ == "__main__":
    main()
```

```python
def main():

    N = int(input().strip())
    
    file_permissions = {}
    
    for _ in range(N):
        line = input().strip().split()
        filename = line[0]
        operations = line[1:]  
        file_permissions[filename] = set(operations)  
    
  
    M = int(input().strip())
    
    for _ in range(M):
        request = input().strip().split()
        operation = request[0]
        filename = request[1]
        
    
        if filename in file_permissions and operation in file_permissions[filename]:
            print("OK")
        else:
            print("Access denied")


if __name__ == "__main__":
    main()
```
